집행일시 열을 이용해서 '연, '월', '일', '연월', '시', '분', '요일' 파생 변수를 만든다.

```python
# to_datetime() 메소드를 사용하면 object(문자열) 타입의 데이터를 datetime 타입의 데이터로 변환할 수 있다.
df['집행일시'] = pd.to_datetime(df.집행일시)
df.dtypes
```

```python
print(type(df.집행일시))
# 연,월,일 ...을 꺼내기 위해 데이터 타입을 바꿔준다.
# 시리즈에서는 str이 먹지 않기 때문에 .str을 해줬었는데 집행 일시는 datetime으로 바꿔서 연,월,일을 꺼낸다. 
print(type(df.집행일시.dt))
print(df.집행일시.dt.year)
print(df.집행일시.dt.year.astype(np.int32))
# print(df.집행일시.dt.year.astype(np.int16)) # 굳이 안 해줘도 되지만 데이터 낭비를 막기 위해 int64 -> int16으로 줄여준다.
```

```python
df['연'] = df.집행일시.dt.year.astype(int)
df['월'] = df.집행일시.dt.month.astype(int)
df['일'] = df.집행일시.dt.day.astype(int)
df['연월'] = df.연.astype(str) + '-' + df.월.astype(str)
df['시'] = df.집행일시.dt.hour.astype(int)
df['분'] = df.집행일시.dt.minute.astype(int)
df['요일'] = df.집행일시.dt.dayofweek.astype(int) # 0(월), 1(화), 2(수), 3(목), 4(금), 5(토), 6(일)
df
```

집행목적 워드클라우드

```python
from wordcloud import WordCloud
```

```python
# 집행목적이 누락값인 데이터를 확인한다.
df[df.집행목적.isnull()]
```

```python
# 집행목적이 누락값인 데이터가 존재하므로 누락값을 공백으로 채워준다.
# drop해서 없애도 되지만 집행장소를 살리기 위해 ''로 바꿨다.

df.집행목적.fillna('', inplace=True)
df[df.집행목적.isnull()]
```

```python
# 파이썬 문자열 메소드 중에서 특정 문자열을 문자열과 문자열에 삽입하는 join() 메소드를 사용해서 집행목적 열에 입력된 데이터를 하나로 연결한다.
# => 워드클라우드로 만들기 위함
data = ' '.join(df.집행목적)
data[:100]
```

```python
wordcloud = WordCloud(
    background_color='white',
    font_path='C:\\Windows\\Fonts\\NanumGothicCoding.ttf',
    relative_scaling=0.5
).generate(data) # .generate_from_frequencies(data) 에서 generate만 남기는 것은 딕셔너리로 안 할 것이기 때문
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

집행장소 워드클라우드

```python
df[df.집행장소.isnull()]
```

```python
# 집행장소가 누락값인 데이터가 존재하므로 누락값을 공백으로 채워준다.
df.집행장소.fillna('', inplace=True)
df[df.집행장소.isnull()]
```

```python
df.head()
```

```python
# 집행장소가 '-'인 데이터를 확인한다.
df[df.집행장소 == '-']
```

```python
# 집행장소가 '-'인 데이터가 존재하므로 '-'을 공백으로 채워준다.
df.loc[df.집행장소 == '-', '집행장소'] = ''
df[df.집행장소 == '-']
```

```python
# 집행장소에 주소가 많이 보이기 때문에 주소는 전처리를 통해 제거하고 상호명이라는 컬럼을 만들어 집행장소에서 상호명을 가져온다.
# 뼈큰감자탕 (대구 중구 명덕로) => () 앞에 있는 게 맛집 상호명!
print('뼈큰감자탕 (대구 중구 명덕로)'.split('(')[0].strip())
df['상호명'] = df.집행장소.apply(lambda x: x.split('(')[0].strip()) # 뼈큰감자탕이 x로 들어가는 것
df.head()
```

```python
# 상호명이 공백인 데이터의 집행목적을 확인한다.
a = df.loc[df.상호명 == '', '집행목적'].value_counts() # .value_counts(): 중복되는 개수 보고 싶을 때
print(a)
# a.to_csv('./data/a.csv')
```

```python
# 상호명이 '-'인 데이터의 집행목적을 확인한다.
df.loc[df.상호명 == '-', '집행목적'].value_counts()
```

```python
# 불필요한 데이터(상호명이 공백 또는 '-'인 데이터)를 제거한다.
df_shop = df[(df.상호명 != '') & (df.상호명 != '-')]
df_shop
```

```python
shop = df_shop.상호명.value_counts()
shop.head(50)
# shop.to_csv('./data/shop.csv')
```

```python
# 같은 상호명이지만 다른 문자열로 표기된 상호명을 하나로 통일한다. => 전처리!
df_shop.loc[df_shop.상호명 == '（주）바닷가작은부엌', '상호명'] = '바닷가작은부엌'
df_shop.loc[df_shop.상호명 == '㈜바닷가작은부엌', '상호명'] = '바닷가작은부엌'
df_shop.loc[df_shop.상호명 == '바닷가작은부엌덕수궁점', '상호명'] = '바닷가작은부엌'

df_shop.loc[df_shop.상호명 == '（주）뚜리삼', '상호명'] = '뚜리삼'
df_shop.loc[df_shop.상호명 == '㈜뚜리삼', '상호명'] = '뚜리삼'
```

```python
top_1000 = df_shop.상호명.value_counts().head(1000)
top_1000 = pd.DataFrame(top_1000).reset_index()
top_1000.columns = ['상호명', '개수']
top_1000
```

```python
wordcloud = WordCloud(
    background_color='white',
    font_path='C:\\Windows\\Fonts\\NanumGothicCoding.ttf',
    relative_scaling=0.5,
    max_font_size=40,
    max_words=1000
).generate(' '.join(df_shop['상호명']))
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
