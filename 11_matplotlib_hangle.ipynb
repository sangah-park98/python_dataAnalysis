```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.font_manager as fm
# 한글 폰트를 사용할 때 마이너스 부호가 깨지는 것을 방지한다.
mpl.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100, 100, 50).cumsum()
data
```

```python
plt.figure(figsize=(8, 6)) # 차트 크기 설정
# plt.rcParams['figure.figsize'] = (8, 6) # 차트 크기 설정
plt.plot(range(50), data, 'b') # range(50, 100)은 x축의 50부터 100까지의 범위를 의미한다.
plt.title('시간별 가격 추이') # 차트 제목
plt.xlabel('시간(분)') # x축 제목
plt.ylabel('가격') # y축 제목
plt.show()
```

한글 폰트를 사용하도록 설정하는 방법은 3가지가 있다.  

(1) FontProperties를 사용하는 방법 => 그래프에 폰트가 필요한 항목마다 개별적으로 지정한다.  
&nbsp;&nbsp;&nbsp;&nbsp;텍스트를 지정하는 항목에 지정해서 사용할 수 있고, 지정해준 항목에만 해당 폰트가 적용된다.  
&nbsp;&nbsp;&nbsp;&nbsp;matplotlib.pyplot => title(), xlabel(), ylabel(), legend(), text()  
&nbsp;&nbsp;&nbsp;&nbsp;matplotlib.axes => set_title()  
(2) matplotlib 라이브러리의 rcParams[]로 전역 글꼴 설정 => 그래프의 폰트 설정이 필요한 파일마다 지정한다.  

(3) 폰트 설정 정보를 matplotlibrc 파일에 직접 넣어준다. => 그래프의 폰트가 필요한 컴퓨터 단위로 지정한다.

```python
font = 'C:\\Windows\\Fonts\\NanumGothicCoding.ttf'
fontprop = fm.FontProperties(fname=font, size=12)
```

```python
plt.figure(figsize=(8, 6)) 
plt.plot(range(50), data, 'b')
plt.title('시간별 가격 추이', fontproperties=fontprop) 
plt.xlabel('시간(분)', fontproperties=fontprop) 
plt.ylabel('가격', fontproperties=fontprop) 
plt.show()
```

```python
# (2) 폰트 기본 설정 정보 읽기
print('기본으로 설정된 폰트 이름:', plt.rcParams['font.family'])
print('기본으로 설정된 폰트 크기:', plt.rcParams['font.size'])
```

```python
# (2-1) 이 방법이 제일 편해!! ----------------------------------
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 12
```

```python
print('변경된 폰트 이름:', plt.rcParams['font.family'])
print('변경된 폰트 크기:', plt.rcParams['font.size'])
```

```python
plt.figure(figsize=(8, 6)) 
plt.plot(range(50), data, 'b')
plt.title('시간별 가격 추이') 
plt.xlabel('시간(분)') 
plt.ylabel('가격') 
plt.show()
```

```python
# (2-2)
# rcParams 대신 FontProperties의 plt.rc를 사용하는 방법이 있다.
font = 'C:\\Windows\\Fonts\\NanumGothicCoding.ttf'
font_name = fm.FontProperties(fname=font, size=12).get_name()
print(font_name)
plt.rc('font', family=font_name)
```

```python
plt.figure(figsize=(8, 6)) 
plt.plot(range(50), data, 'b')
plt.title('시간별 가격 추이') 
plt.xlabel('시간(분)') 
plt.ylabel('가격') 
plt.show()
```

(3) 폰트 설정 정보를 matplotlibrc 파일에 직접 넣어준다. => 그래프의 폰트가 필요한 컴퓨터 단위로 지정한다.

```python
# mpl.matplotlib_fname()를 실행하면 출력되는 경로의 matplotlibrc 파일을 수정한다.
# matplotlibrc 파일을 수정하면 쥬피터 노트북을 실행할 때마다 별도의 설정 없이 matplotlibrc 파일의 설정을 사용한다.
print(mpl.matplotlib_fname())
# ont.family: sans-serif => NanumGothicCoding
# font.size: 10.0 => 15.0
```
