워드클라우드를 그레이스케일로 표시하는 함수를 만든다.

```python
import random
```

```python
# word: 워드클라우드로 표시할 단어
# font_size: 워드클라우드로 표시할 단어의 크기
# position: 워드클라우드로 표시할 단어의 위치(좌표) => ()에 묶여 튜플 형태로 나온다.
# orientation: 워드클라우드로 표시할 단어의 방향 (마음대로 바뀐다.)
# random_state: 숫자를 넘겨주면 워드클라우드로 표시되는 단어들의 위치가 고정된다.
# hwargs: 워드클라우드로 표시할 단어들의 글꼴이 저장된 경로

def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs):
    # print(word, font_size, position, orientation, random_state, hwargs)
    return 'hsl(0, 0%%, %d%%)' % random.randint(30, 60)
    # return 'hsl(%d, %d%%, %d%%)' % (random.randint(0, 255), random.randint(0, 100), random.randint(30, 60)) # alice가 칼라로 나온다.
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(
    wordcloud.recolor(color_func=gray_color_func, random_state=200), 
    interpolation='bilinear'
)
plt.axis('off')
plt.show()
```

```python
# 워드 클라우드 작성 시 반드시 필요한 최소식
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
from wordcloud import STOPWORDS
```

```python
# text = open('./data/ebook.txt').read()
# UnicodeDecodeError: 'cp949' codec can't decode byte 0xe2 in position ~~~~ 에러가 발생되면 
text = open('./data/ebook.txt', encoding='UTF-8').read()

# print(text)
mask = np.array(Image.open('./images/stormtrooper_mask.png'))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(max_words=500, background_color='white', stopwords=stopwords, mask=mask).generate(text)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=200), interpolation='bilinear')
plt.axis('off')
plt.show()
```

워드클라우드의 마스크로 컬러 이미지를 사용할 경우 마스크로 사용한 이미지의 색상에 가깝게 워드클라우드 단어의 색상을 지정할 수 있다.

```python
from wordcloud import ImageColorGenerator
```

```python
image_colors = ImageColorGenerator(mask)
```

```python
plt.figure(figsize=(15, 10))
# ImageColorGenerator 클래스 객체를 생성할 때 생성자로 넘겨준 mask 이미지의 객체를 recolor() 메소드의
# color_func 속성에 지정한다.
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
